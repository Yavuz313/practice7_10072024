**1. Öğrenci Not Sistemi**

**Senaryo**

Bir öğretmen, öğrencilerinin ders notlarını ve durumlarını daha verimli bir şekilde takip edebilmek için bir yazılım geliştirmek istiyor. Bu yazılım, öğrencilerin adlarını ve üç ders için aldıkları notları saklamalı, notların ortalamasını hesaplamalı ve öğrencinin durumunu belirlemelidir. Öğretmen, öğrencilerin durumlarını kolayca görüntüleyebilmek ve gerektiğinde yeni öğrenciler ekleyebilmek istemektedir.

**Sorun**

Manuel olarak öğrencilerin notlarını ve durumlarını takip etmek, hatalara açık ve zaman alıcı olabilir. Bu nedenle, öğrencilerin notlarını saklayacak, ortalamalarını hesaplayacak ve durumlarını belirleyecek bir yazılım çözümüne ihtiyaç vardır. Bu çözüm, öğretmenin işini kolaylaştıracak ve öğrencilerin durumlarını daha doğru bir şekilde değerlendirmesini sağlayacaktır.

**Talimatlar**

    1. Öğrenci Ekleme:
        Kullanıcıdan öğrencinin adını ve üç ders için aldığı notları girmesini isteyin.
        Notların geçerli sayılar olup olmadığını kontrol edin ve geçersiz girişler için uyarı verin.
        Öğrencinin adı ve notları, öğrenci listesine eklenmelidir.

    2. Ortalama Hesaplama:
        Her öğrencinin notlarının ortalamasını hesaplayın.

    3. Durum Belirleme:
        Öğrencinin ortalamasına göre durumunu belirleyin:
            Ortalama 90 ve üzeri ise "Mükemmel"
            Ortalama 75 ve üzeri ise "İyi"
            Ortalama 50 ve üzeri ise "Geçti"
            Ortalama 50'nin altı ise "Kaldı"

    4. Öğrenci Durumlarını Görüntüleme:
        Tüm öğrencilerin adlarını, not ortalamalarını ve durumlarını listeleyin.

    5. Çıkış:
        Kullanıcıya programdan çıkma seçeneği sunun.

**Program Akışı**

* Kullanıcıya ana menüyü görüntüleyin ve seçenekleri sunun.
* Kullanıcı 1'i seçerse, yeni bir öğrenci eklemek için öğrencinin adını ve notlarını alın.
* Kullanıcı 2'yi seçerse, tüm öğrencilerin durumlarını listeleyin.
* Kullanıcı 3'ü seçerse, programdan çıkın.
* Kullanıcı geçersiz bir seçim yaparsa, tekrar menüyü görüntüleyin.

**Örnek Kullanıcı Etkileşimi**

```

1. Öğrenci Ekle
2. Öğrenci Durumlarını Görüntüle
3. Çıkış

Seçiminiz (1-3): 1
Öğrencinin adını girin: Ali
1. dersin notunu girin: 85
2. dersin notunu girin: 90
3. dersin notunu girin: 78
Ali isimli öğrenci eklendi.

1. Öğrenci Ekle
2. Öğrenci Durumlarını Görüntüle
3. Çıkış

Seçiminiz (1-3): 2
Ali: Ortalama - 84.33, Durum - İyi

1. Öğrenci Ekle
2. Öğrenci Durumlarını Görüntüle
3. Çıkış

Seçiminiz (1-3): 3
Programdan çıkılıyor...

```

````python
def add_student(students):
    name = input("Öğrencinin adını giriniz: ")
    grades = []
    
    for i in range(1, 4):
        grade = input(f"{i}. dersin notunu giriniz: ")
        if grade.isdigit():
            grades.append(int(grade))
        else:
            print("Geçersiz not girdiniz. Lütfen tekrar deneyin.")
            return
        
    students[name] = grades
    
    print(f"{name} ismli öğrenci eklendi.")
    
def calculate_average(grades):
    return sum(grades) / len(grades)

def determine_status(average):
    if average >= 90:
        return "Mükemmel"
    elif average >= 75:
        return "İyi"
    elif average >= 50:
        return "Geçti"
    else:
        return "Kaldı"
    
    
def display_student_status(students):
    for name, grades in students.items():
        average = calculate_average(grades)
        status = determine_status(average)
        print(f"{name}: Ortalama - {average:.2f}, Durum - {status}")
        
        
def main():
    students = {}
    while True:
        print("""
              1. Öğrenci Ekle
              2. Öğrenci Durumlarını Görüntüle
              3. Çıkış
              """)
        
        choice = input("Lütfen bir seçim yapınız (1-3): ")
        
        if choice == "1":
            add_student(students)
            
        elif choice == "2":
            display_student_status(students)
            
        elif choice == "3":
            print("Programdan çıkılıyor...")
            break
        else:
            print("Geçersiz seçim, lütfen tekrar deneyiniz.")
            
main()
````

**2. Yaş Sınıflandırma Sistemi**

**Sorun**:

Bir sosyal kulüp, üyelerinin yaşlarını düzenli olarak takip etmek ve belirli yaş aralıklarında olan kişileri etkinlikler için daha kolay sınıflandırmak istiyor. Ancak, manuel olarak yaşları takip etmek ve sınıflandırmak zaman alıcı ve hataya açık olabilir. Bu nedenle, yaşları ekleme, listeleme ve belirli yaş aralıklarındaki kişileri filtreleme işlemlerini otomatikleştiren bir yazılım çözümüne ihtiyaç duyulmaktadır.

**Talimatlar**:

    1. Yaş Ekleme:
        Kullanıcıdan bir yaş girişi alınmalı ve geçerli bir sayı olup olmadığı kontrol edilmelidir.
        Geçerli bir yaş girildiğinde, bu yaş listeye eklenmelidir.
        Geçersiz bir yaş girildiğinde, kullanıcıya hata mesajı gösterilmelidir.

    2. Tüm Yaşları Listeleme:
        Eklenmiş tüm yaşları listeleyin.
        Eğer henüz hiçbir yaş eklenmediyse, kullanıcıya uygun bir mesaj gösterilmelidir.

    3. Belirli Yaş Aralığındaki İnsanları Filtreleme:
        Kullanıcıdan minimum ve maksimum yaş aralıklarını alarak belirli bir yaş aralığında olan kişileri filtreleyin.
        Filtrelenen yaşlar listelenmelidir.
        Eğer belirlenen yaş aralığında kimse yoksa, kullanıcıya uygun bir mesaj gösterilmelidir.

    4. Çıkış:
        Kullanıcı programdan çıkmak istediğinde, program sonlandırılmalıdır.

**Kullanıcı Etkileşimi Örneği:**

1. Yaş Ekleme:
* Kullanıcı "1" seçeneğini seçer ve yaş girer. Geçerli bir yaş girildiğinde, yaş listeye eklenir.

```
Yaşı girin: 25
Yaş eklendi: 25
```

2. Tüm Yaşları Listeleme:

* Kullanıcı "2" seçeneğini seçer ve eklenmiş tüm yaşları görür.

```
Yaşlar:
- 25
```

3. Belirli Yaş Aralığındaki İnsanları Filtreleme:

* Kullanıcı "3" seçeneğini seçer ve minimum ve maksimum yaş aralığını girer. Belirlenen yaş aralığında olan kişiler listelenir.

```
Minimum yaşı girin: 20
Maksimum yaşı girin: 30
20-30 yaş aralığındaki insanlar:
- 25
```

4. Çıkış:

* Kullanıcı "4" seçeneğini seçer ve programdan çıkar.

```
Programdan çıkılıyor...
```

````python
age_list = []

def add_age():
    age = input("Yaşı giriniz: ")
    if age.isdigit():
        age_list.append(int(age))
        print(f"Yaş eklendi: {age}")
    else:
        print("Geçersiz bir yaş girdiniz. Lütfen tekrar deneyin.")
        
def list_ages():
    if not age_list:
        print("Henüz yaş eklenmedi.")
        
    else:
        print("Yaşlar: ")
        for age in age_list:
            print(f"- {age}")
            
def filter_ages(min_age, max_age):
    filtered_ages = list(filter(lambda age: min_age <= age <= max_age, age_list))
    
    if not filtered_ages:
        print(f"{min_age}-{max_age} yaş aralığında insan yok.")
        
    else:
        print(f"{min_age}-{max_age} yaş aralığındaki insanlar:")
        for age in filtered_ages:
            print(f"- {age}")
            
            
def main():
    while True:
        print("""
              Yaş Sınıflandırma Sistemi
              -------------------------
              1. Yaş Ekle
              2. Yaşları Listele
              3. Belirli Yaş Aralığındaki İnsanları Filtrele
              4. Çıkış
              """)
        
        choice = input("Seçiminiz (1-4): ")
        
        if choice == "1":
            add_age()
            
        elif choice == "2":
            list_ages()
            
        elif choice == "3":
            min_age = input("Minimum yaşı giriniz:")
            max_age = input("Maksimum yaşı giriniz:")
            
            if min_age.isdigit() and max_age.isdigit():
                filter_ages(int(min_age), int(max_age))
            else:
                print("Geçersiz yaş aralığı girdiniz. Lütfen tekrar deneyin.")
                
        elif choice == "4":
            print("Programdan çıkılıyor...")
            break
        
        else:
            print("Geçersiz seçim, lütfen tekrar deneyin")
            
main()
````

**3. Kelime Tahmin Oyunu**

**Sorun**:

Bir grup arkadaş eğlenceli bir oyun oynamak istiyor. Oyunda bir kişi bir kelimeyi karıştırarak arkadaşlarına gösteriyor ve onların bu karışık harflerden doğru kelimeyi tahmin etmelerini istiyor. Ancak manuel olarak kelime seçip harfleri karıştırmak ve tahminleri takip etmek zor olabiliyor. Bu işlemi otomatik hale getiren bir program yazmak istiyorlar.

**Senaryo**:

Bu program, bilgisayarın rastgele bir kelime seçip harflerini karıştırarak oyuncuya göstermesini sağlar. Oyuncu, verilen harflerden doğru kelimeyi tahmin etmeye çalışır. Oyuncuya üç deneme hakkı verilir ve eğer doğru kelimeyi tahmin edemezse oyun sona erer ve doğru kelime gösterilir.

**Talimatlar**:

* Program çalıştığında, oyuncuya karışık harfler gösterilir ve doğru kelimeyi tahmin etmesi istenir.
* Oyuncunun üç deneme hakkı vardır.
* Oyuncu doğru kelimeyi tahmin ederse, oyun sona erer ve oyuncu tebrik edilir.
* Oyuncu üç deneme hakkını da kullanıp doğru kelimeyi tahmin edemezse, oyun sona erer ve doğru kelime gösterilir.

**Kullanıcı Etkileşimi Örneği:**

1. Başlangıç:

```
Kelime Tahmin Oyununa Hoşgeldiniz!
Karışık harfler: lmua
```

2. Tahminler:

* Oyuncu tahminini girer ve doğru olup olmadığını öğrenir.

```
Tahmininizi girin (Kalan deneme hakkı: 3): ulam
Yanlış tahmin. Tekrar deneyin.

Tahmininizi girin (Kalan deneme hakkı: 2): alma
Yanlış tahmin. Tekrar deneyin.

Tahmininizi girin (Kalan deneme hakkı: 1): elma
Tebrikler! Doğru tahmin ettiniz: elma
```

3. Oyunu Kaybetme:

* Oyuncu 3 deneme sonunda doğru kelimeyi bulamazsa, doğru kelime gösterilir.

```
Tahmininizi girin (Kalan deneme hakkı: 3): ulam
Yanlış tahmin. Tekrar deneyin.

Tahmininizi girin (Kalan deneme hakkı: 2): alma
Yanlış tahmin. Tekrar deneyin.

Tahmininizi girin (Kalan deneme hakkı: 1): maul
Yanlış tahmin. Tekrar deneyin.

Maalesef, deneme haklarınız bitti. Doğru kelime: elma
```

````python
import random

words = ["elma", "armut", "kiraz", "muz", "cilek"]

def get_scrambled_word(word_list):
    word = random.choice(word_list)
    scrambled = "".join(random.sample(word, len(word))) # kelimenin uzunluğu kadar başka bir kelimeyi harfleri karıştırarak ve birleştirerek oluşturacak
    return word, scrambled

def play_game():
    word, scrambled_word = get_scrambled_word(words)
    attempts = 3
    
    print("Kelime Tahmin Oyununa Hoşgeldiniz!")
    print(f"Karışık Harfler: {scrambled_word}")
    
    while attempts > 0:
        tahmin = input(f"Tahmininizi giriniz (Kalan deneme hakı: {attempts}): ")
        
        if tahmin == word:
            print(f"Tebrikler! Doğru tahmin ettiniz: {word}")
            break
            
        else:
            print("Yanlış tahmin. Tekrar deneyin.")
            attempts -= 1
            
    if attempts == 0:
        print(f"Maalesef, deneme hakkınız bitti. Doğru kelime: {word}")
            
            
play_game()
````

**4. Çalışan Maaş Yönetim Sistemi** ***----ÖDEV----***
